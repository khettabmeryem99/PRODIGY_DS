```python
#liste of contries :
countries = ['Morocco', 'Canada', 'United Kingdom', 'China', 'Germany', 'Mexico' , 'Australia']
#Filtering the data 
data_filtered = data[data['Country Name'].isin(countries)]

#Creating the bar chart
plt.bar (data_filtered['Country Name'], data_filtered['2023'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2023')
plt.title('Population of selected countries in 2023')
plt.xticks(rotation=40, ha='right') 
plt.show()
```
